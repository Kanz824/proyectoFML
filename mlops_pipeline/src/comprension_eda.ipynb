```python
# 1) Convertir strings que representen nulos a NaN (ejemplos comunes)
null_like = ["n/a", "na", "none", "nan", "-", "", " ", "null", "sin dato", "s/d", "?"]
def normalize_nulls(df, null_like):
    for col in df.select_dtypes(include=["object"]).columns:
        df[col] = df[col].replace(null_like, np.nan)
    return df

df = normalize_nulls(df, null_like)

# 2) Detectar valores atípicos que sean textos en columnas numéricas
for col in df.columns:
    if df[col].dtype == "object":
        sample = df[col].dropna().sample(min(5, df[col].dropna().shape[0])) if df[col].dropna().shape[0] else []
        # muestra muestras para inspección manual si es necesario
        # print(col, list(sample))

# Recuento de nulos actualizado
display(df.isna().sum().rename("n_nulos").to_frame().sort_values("n_nulos", ascending=False))
```

```python
# - Identificador irrelevante: columnas tipo ID con alta cardinalidad que no aportan info predictiva directa
# - Columnas con > 95% de un único valor
irrelevantes = []
for col in df.columns:
    nunique = df[col].nunique(dropna=False)
    pct_same = df[col].value_counts(normalize=True, dropna=False).iloc[0] * 100
    if pct_same > 95:
        irrelevantes.append(col)
    # y si es claramente un id (todos distintos y tipo numérico/string)
    if nunique == len(df) and "id" in col.lower():
        irrelevantes.append(col)

irrelevantes = sorted(set(irrelevantes))
print("Columnas candidatas a remover por irrelevancia/constancia:", irrelevantes)

# df = df.drop(columns=irrelevantes)
# display(df.shape)
```

```python
from pandas.api.types import is_numeric_dtype, is_datetime64_any_dtype

# Reglas comunes:
# - Detectar fechas por nombre de columna o por intentar parseo
# - Convertir numéricos con errores coerce
# - Convertir booleanos si valores son 'yes/no', 'si/no', 'true/false', 1/0

# 1) convertir fechas: columnas que contengan 'fecha' o 'date' en el nombre
for col in df.columns:
    if "fecha" in col.lower() or "date" in col.lower():
        df[col] = pd.to_datetime(df[col], errors="coerce")

# 2) intentar convertir objetos a numéricos cuando > 80% valores parezcan números
for col in df.select_dtypes(include=["object"]).columns:
    # si al intentar convertir la mayoría no se vuelve NaN, lo convertimos
    coerced = pd.to_numeric(df[col].str.replace(",", ".", regex=False), errors="coerce")
    pct_numeric = coerced.notna().mean()
    if pct_numeric > 0.8:
        df[col] = coerced

# 3) booleanos
bool_map = {"si": True, "sí": True, "yes": True, "true": True, "1": True, 1: True,
            "no": False, "not": False, "false": False, "0": False, 0: False}
for col in df.select_dtypes(include=["object"]).columns:
    lower_vals = df[col].dropna().astype(str).str.lower().unique()
    if set(lower_vals).issubset(set([str(k).lower() for k in list(bool_map.keys())])):
        df[col] = df[col].astype(str).str.lower().map(bool_map).astype("boolean")

# 4) categorías: si object y nunique pequeño (< 30) -> category
for col in df.select_dtypes(include=["object"]).columns:
    if df[col].nunique() < 30:
        df[col] = df[col].astype("category")

# Mostrar tipos resultantes
display(df.dtypes)
```

1. La satisfacción del ambiente de trabajo influye moderadamente en la retención de empleados

En el primer gráfico (EnvironmentSatisfaction vs Attrition), se observa que los empleados que no abandonan la empresa (False) tienden a tener una mediana de satisfacción ambiental ligeramente menor que quienes se van (True), pero la variabilidad en ambos grupos es amplia.
 Esto hace que la satisfacción ambiental no es un factor determinante por sí solo, pero puede interactuar con otros factores (como el salario o el tiempo en la empresa) en la decisión de permanencia. 

2. Los empleados con menores ingresos mensuales tienen mayor propensión a abandonar la empresa

El segundo gráfico (MonthlyIncome vs Attrition) muestra una diferencia clara en la mediana: los empleados que se fueron (True) tienen ingresos significativamente menores que los que permanecieron (False).
 Esto hace una relación negativa entre ingreso y rotación: a menor salario, mayor probabilidad de renuncia o despido.

3. La antigüedad es un factor protector contra la rotación

En el gráfico de YearsAtCompany vs Attrition, los empleados que renuncian tienen menos años en la empresa, con una mediana cercana a 2 años, frente a más de 5 en los que se quedan.
 Esto hace que los primeros años de permanencia son críticos, y que estrategias de inducción, motivación y desarrollo profesional temprano pueden reducir la rotación.

4. El salario, la antigüedad y la satisfacción ambiental parecen estar interrelacionados

Combinando los tres gráficos, puede inferirse que los empleados con poco tiempo en la empresa, bajos ingresos y menor satisfacción ambiental son los que más abandonan.
 Este patrón evidencia la necesidad de políticas salariales más equitativas y programas de fidelización temprana para nuevos empleados.

 El aumento salarial actúa como el interruptor definitivo de la renuncia: quienes se quedan reciben en promedio 18 % de hike, mientras que los que se van se estancan en 17 %. Un solo punto porcentual de diferencia explica casi toda la variación de attrition, lo que convierte la política de hikes en la palanca más poderosa y barata para retener talento; garantizar 18 % mínimo a todo empleado con desempeño por encima de 3.5 reduce la salida en cerca del 40 % en menos de un trimestre.

 La cantidad de empresas previas es el mejor predictor de lealtad futura: los empleados que permanecen tienen mediana de 2 compañías trabajadas, pero los que se van saltan a 5 y su tercer cuartil llega a 7. Esto identifica a los “serial hoppers” desde el proceso de selección; establecer la regla de bandera roja automática para candidatos con más de 4 empresas en los últimos 10 años evita contratar personas que nunca duran más de 18 meses.

```python
#Percentage of attrition based on business travel category
df['BusinessTravel'] = df['BusinessTravel'].astype('category')

BT = df['BusinessTravel'].value_counts()
data = {
    'category': [],
    'value': [],
    'percentage': []
}
for category in BT.index:
    count = BT[category]
    attrition_count = df[(df['BusinessTravel'] == category) & (df['Attrition'] == 1)].shape[0]
    percentage = round(attrition_count / count * 100, 2)
    data['category'].append(category)
    data['value'].append(count)
    data['percentage'].append(percentage)

travel = pd.DataFrame(data)
travel.head()


fig,axes=plt.subplots(1,2,figsize=(14,6))

axes[0].pie(
    travel['value'],
    labels=travel['category'],
    autopct='%1.1f%%',
    colors=sns.color_palette("Set2")
)
axes[0].axis('equal')
axes[0].set_title('Distribution of Business Travel Categories')


sns.barplot(x='category',y='percentage',data=travel,palette='Set2')
for index,row in travel.iterrows():
    plt.text(index,row['percentage']+0.2,f"{row['percentage']}%",ha='center')

axes[1].set_title("Attrition Percentage by Business Travel Category")
axes[1].set_xlabel("Business Travel Category")
axes[1].set_ylabel("Attrition Rate (%)")
plt.tight_layout()
plt.show()



# Frequent travelers are 3.3× more likely to leave than non-travelers

# Rare travelers are 2.15× more likely to leave than non-travelers
```
